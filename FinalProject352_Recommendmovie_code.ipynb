# **Extracting genres**
ใช้ไลบรารี 'ast' เพื่อแปลง str เป็น list เพราะตอนนี้ค่าทั้งหมดเป็น str


```python
import ast
# function to get a list of dict values
def get_dict_val(data):
    ls = []
    c =ast.literal_eval(data)
    for i in c:
        ls.append(i['name'])
    if len(ls)>0:
        return ls
    else: return ['Missing']
```

# **Extracting cast name**
เพราะต้องการเฉพาะชื่อ cast name

```python
df_mov['cast']
```

```python
def get_cast(data):
    ls = []
    c =ast.literal_eval(data)
    for i in c:
        if len(ls)<3:
            ls.append(i['name'])
    if len(ls)>0:
        return ls
    else: return ['Missing']
```

```python
df_mov['cast']=df_mov['cast'].apply(get_cast)
```

```python
df_mov['cast']
```

```python
df_mov['crew']
```

```python
def get_crew(data,dept):
    ls = []
    c =ast.literal_eval(data)
    for i in c:
        if i['job'] == dept:
            ls.append(i['name'])
    if len(ls)>0: return ls
    else: return ['Missing']
```

```python
df_mov['director']=df_mov['crew'].apply(lambda x: get_crew(x,'Director'))
df_mov['producer']=df_mov['crew'].apply(lambda x: get_crew(x,'Producer'))
df_mov.drop(columns='crew',inplace=True)
```

### **Data Preprocessing for Model**

```python
df_mov = df_mov[['movie_id','title','genres','keywords','overview','cast','director']]
df_mov
```

```python
df_mov.reset_index(inplace=True)
# converting string to list
df_mov['overview']=df_mov['overview'].str.split()
# remove space between words from a list
df_mov['genres']=df_mov['genres'].apply(lambda x:[ i.replace(' ','') for i in x])
df_mov['keywords']=df_mov['keywords'].apply(lambda x:[ i.replace(' ','') for i in x])
df_mov['cast']=df_mov['cast'].apply(lambda x:[ i.replace(' ','') for i in x])
df_mov['director']=df_mov['director'].apply(lambda x:[ i.replace(' ','') for i in x])
df_mov
```

```python
# creating a tag column combining relative columns
df_mov['tags']=df_mov['genres']+df_mov['keywords']+df_mov['overview']+df_mov['cast']+df_mov['director']
```

### **Steming** 
Stemming คือ กระบวนตัดส่วนท้ายของคำ ประโยคที่มีความหมายเหมือนกันในขณะที่ทำงานกับ ML 

```python
import nltk
from nltk.corpus import stopwords
import string
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
ps.stem('loved'),ps.stem('flies'),ps.stem('actors')
```

```python
def text(text):
    ls = []
    for i in text.split():
        ls.append(ps.stem(i))
    return " ".join(ls)
```

```python
text('loved loving loves danced dancing')
```

```python
text('action adventure fantasy sciencefiction')
```

```python
new['tags']=new['tags'].apply(text)
```

# **Removing Stop Words:**
คำหยุดคือคำที่ใช้ในการสร้างประโยค เช่น - is, am,are,and, if 

```python
#ลดมิติข้อมูลโดยใช้ CountVectorizer
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

# **recommended movies**

```python
def recommend(movie):
    movie_index = new[new['title']==movie].index[0] # รับชื่อภาพยนต์
    distances = similarity[movie_index] # รับข้อมูล similarity ขอภาพยนต์ทุกเรื่อง
    # sorting เอาข้อมูล 5 อันดับแรก
    movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    for i in movies_list:
        print(new.iloc[i[0]].title)
```
